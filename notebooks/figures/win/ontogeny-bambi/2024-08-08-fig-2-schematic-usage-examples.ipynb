```python
import numpy as np
import pandas as pd
import seaborn as sns
from aging.plotting import format_plots, PLOT_CONFIG, save_factory, figure, COLORMAPS
```

```python
format_plots()
saver = save_factory(PLOT_CONFIG.save_path / "fig-2-lab-comments", tight_layout=False)
```

```python
keep_syllables = np.loadtxt('/n/groups/datta/win/longtogeny/data/ontogeny/version_11/to_keep_syllables_raw.txt', dtype=int)
df = pd.read_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_11-1/ontogeny_males_raw_usage_matrix_v00.parquet')[keep_syllables]
```

```python
x = df.query('age < 16').mean()
x = x.sort_values(ascending=False)
```

```python
fig = figure(0.6, 0.5)
ax = fig.gca()
ax.plot(x.values, '-', color=COLORMAPS.ont_male(255))
ax.set(xlabel="Syllable ID", xticks=[], ylabel="Syllable usage", yticks=[])
sns.despine()
saver(fig, "young-male-example-schematic");
```

```python
sorting = x.index
```

```python
y = df.query('age > 16 & age < 26').mean().loc[sorting]
```

```python
fig = figure(0.6, 0.5)
ax = fig.gca()
ax.plot(y.values, '-', color=COLORMAPS.ont_male(255))
ax.set(xlabel="Syllable ID", xticks=[], ylabel="Syllable usage", yticks=[])
sns.despine()
saver(fig, "mid-male-example-schematic");
```

```python
z = df.query('age > 85 & age < 110').mean().loc[sorting]
```

```python
fig = figure(0.6, 0.5)
ax = fig.gca()
ax.plot(z.values, '-', color=COLORMAPS.ont_male(255))
ax.set(xlabel="Syllable ID", xticks=[], ylabel="Syllable usage", yticks=[])
sns.despine()
saver(fig, "old-male-example-schematic");
```
