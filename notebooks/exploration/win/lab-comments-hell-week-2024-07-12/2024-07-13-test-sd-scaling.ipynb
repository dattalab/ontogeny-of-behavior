```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

σ = sqrt(ln(1 + (σ^2/μ^2))) 

```python
def scale_sd(sd, mu):
    return np.sqrt(np.log(1 + (np.square(sd) / np.square(mu))))
```

```python
sd = 3
```

```python
mus = np.linspace(0, 5, 25)[1:]
```

```python
log_mus = np.log(mus)
```

```python
plt.plot(log_mus, log_mus, '.')
```

```python
samples = np.random.normal(size=20)
```

```python
new_samples = (samples[:, None] + log_mus)
```

```python
for row in new_samples:
    plt.plot(mus, np.exp(row), '.')
```

```python
scaled_samples = []
for mu in mus:
    s = (samples * scale_sd(1, mu)) + np.log(mu)
    scaled_samples.append(s)
scaled_samples = np.array(scaled_samples).T
```

```python
for row in scaled_samples:
    plt.plot(mus, np.exp(row), '.')
```
