```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

## Females

```python
df = pd.read_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_12/ontogeny_females_raw_usage_matrix_v00.parquet')
```

```python
scores = PCA(n_components=2).fit_transform(df)
```

```python
plt.figure(figsize=(2, 2))
plt.scatter(scores[:, 0], scores[:, 1], c=df.index.get_level_values('age'))
```

```python
pd.Series(scores[:, 0]).corr(pd.Series(df.index.get_level_values('age')))
```

```python
pd.Series(scores[:, 1]).corr(pd.Series(df.index.get_level_values('age')))
```
