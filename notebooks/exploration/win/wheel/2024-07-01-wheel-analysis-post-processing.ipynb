```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from aging.plotting import format_plots
```

```python
format_plots()
```

```python
df = pd.read_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_11-1/wheel_raw_usage_matrix_v00.parquet')
```

```python
ages = df.index.get_level_values('age')
week_floor = np.round(ages).astype(int)
df = df.reset_index(level='age')
df['age'] = week_floor
df = df.set_index('age', append=True)
```

```python
plt.plot(week_floor)
```

```python
df.head()
```

```python
df.index.get_level_values('mouse').unique()
```

```python
df.index.get_level_values('mouse').nunique()
```

```python
df.loc[df.index.get_level_values('mouse') == '12wc3_0']
```

```python
mice = pd.Series(df.index.get_level_values('mouse'))
df = df.reset_index('mouse')
mice.loc[mice == '12wc3_0'] = '12c3_04'
df['mouse'] = mice.to_numpy()
df = df.set_index('mouse', append=True)
```

```python
df.head()
```

```python
drop_values = [13, 79,90]
# Dropping rows with specific 'age' level values
df = df.drop(index=drop_values, level='age')
```

```python
plt.imshow(df.T, aspect='auto', vmax=0.1)
plt.colorbar()
```

```python
mice = pd.Series(df.index.get_level_values('mouse'))
```

```python
control_mask = mice.str.contains('c')
wheel_mask = mice.str.contains('w')

young_mask = mice.str.startswith('12')
old_mask = mice.str.startswith('78')
```

```python
drop_mask = young_mask.to_numpy() & (df.index.get_level_values('age') > 50)
```

```python
keep_mask = ~drop_mask
```

```python
df['experiment_type'] = 'control'
df.loc[wheel_mask.to_numpy(), 'experiment_type'] = 'wheel'

df['age_group'] = 'young'
df.loc[old_mask.to_numpy(), 'age_group'] = 'old'
```

```python
df = df[keep_mask].copy()
```

```python
df = df.set_index(['experiment_type', 'age_group'], append=True)
```

```python
df.to_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_11-1/wheel_raw_usage_matrix_v00_post_processing.parquet')
```
