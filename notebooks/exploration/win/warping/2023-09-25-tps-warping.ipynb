```python
def compute_frame_avg(path, thresh=9, rears=False):
    with h5py.File(path, 'r') as h5f:
        frames = h5f['frames'][()]

    peaks = np.quantile(frames.reshape(len(frames), -1), 0.95, axis=1)
    peaks -= np.median(peaks)
    if rears:
        med_frame = np.median(frames[peaks > thresh], axis=0)
    else:
        med_frame = np.median(frames[peaks < thresh], axis=0)
    return med_frame


def get_pos(idx, source, dest, mapping):
    try:
        return dest[mapping[idx]]
    except KeyError:
        return source[idx]


def map_points(p1, p2):
    src = np.array(p1.all_points)
    try:
        distances = cdist(p1.all_points, p2.all_points)
        dest_map = dict(map(reversed, enumerate(distances.argmin(axis=0))))
        src = np.array(p1.all_points)

        dest = np.array(p2.all_points)
        dest = np.array([get_pos(idx, src, dest, dest_map) for idx in range(len(src))])
        return src, dest
    except ValueError:
        return src, src


def warp(src, dest, img):
    src = torch.tensor(src.copy()).unsqueeze(0).float()
    dest = torch.tensor(dest.copy()).unsqueeze(0).float()
    src = src / img.shape[0] - 0.5
    dest = dest / img.shape[0] - 0.5
    kernel_weights, affine_weights = get_tps_transform(dest, src)

    img = torch.tensor(img, dtype=torch.float)
    img = img.view(1, 1, *img.shape)
    out = warp_image_tps(img, src, kernel_weights, affine_weights)
    return out.squeeze().detach().numpy()
```

```python
class pp_factory:

    def __init__(self, color="red"):
        # set up control points
        self.all_points = [(0, 0), (0, 80), (80, 0), (80, 80), (0, 40), (40, 0), (40, 80), (80, 40)]
        self.color = color

    def plot_point(self, x, y):
        if x is None or y is None:
            return hv.Points(self.all_points).opts(size=1)
        self.all_points.append((x, y))
        return hv.Points(self.all_points).opts(color=self.color, size=10)
```

```python
def mouse_iter(mouse_dict, n_mice):
    rng = random.Random(0)
    for k, v in mouse_dict.items():
        rng.shuffle(v)
        items = v[:n_mice]
        for item in items:
            yield k, item


class AggregateWarpPoints:
    def __init__(self, mouse_dict, baselines, thresh=9, rears=False):
        self.points = defaultdict(list)
        self.rears = rears
        self.thresh = thresh
        self.baselines = baselines
        self.baseline_cycle = cycle(baselines)
        self.n_mice = len(baselines)
        self.mice = mouse_dict
        self.current_pp = pp_factory(color="cyan")
        self.comparator = next(self.baseline_cycle)

        self.mouse_iter = mouse_iter(mouse_dict, self.n_mice)
        self.age, self.file = next(self.mouse_iter)
        self.img = compute_frame_avg(self.file, thresh=self.thresh, rears=self.rears)
        self.interactive_img = None

    def next(self):
        src, target = map_points(self.comparator, self.current_pp)
        self.points[self.age].append((src, target))

        self.age, self.file = next(self.mouse_iter)
        self.current_pp = pp_factory(color="cyan")
        self.comparator = next(self.baseline_cycle)
        self.img = compute_frame_avg(self.file, thresh=self.thresh, rears=self.rears)

    def plot(self, x=None, y=None, value=None):
        if value:
            self.next()
        img = hv.Image(self.img, bounds=(0, 0, 80, 80)).opts(
            cmap="cubehelix", invert_yaxis=True
        )
        img = img * hv.Points(self.comparator.all_points).opts(
            color=self.comparator.color, size=5
        )
        if not value or value is None:
            img = img * self.current_pp.plot_point(x, y).opts(size=5)
        self.interactive_img = img
        return self.interactive_img
```

## Generation of "principal warps"

Warp 3-month old mice into other ages

Steps:
- select a subset of 3 month sessions
- pre-define warp points
- loop through different ages
- within each age, select a subset of sessions
- use a defined 3-month warp point group as a guide to create warp points for each session
- map all 3 month points to all aged points
- save the mappings into a dictionary where the keys are the age map
- repeat for rearing
- repeat for scrunching

```python
random.seed(0)
baseline = []

n_mice = 4
base_mice = grouped['3months_19042021'].copy()

random.shuffle(base_mice)

images = []
for animal in base_mice[:n_mice]:
    young_med_frame = compute_frame_avg(animal)
    img = hv.Image(young_med_frame, bounds=(0, 0, 80, 80)).opts(cmap='cubehelix', invert_yaxis=True)
    pp = pp_factory()
    dm = hv.DynamicMap(pp.plot_point, streams=[hv.streams.Tap(source=img)])
    images.append(img * dm)
    baseline.append(pp)
reduce(add, images)
```

```python
joblib.dump(baseline, '/n/scratch3/users/w/wg41/baseline-warp-points.p')
```

```python
agger = AggregateWarpPoints(dissoc(grouped, "3months_19042021"), baseline)

btn = pn.widgets.Button(
    name="next", button_type="primary", button_style="outline", width=425
)
dm = hv.DynamicMap(
    agger.plot, streams=[hv.streams.Tap(source=agger.interactive_img), btn.param.value]
).opts(width=425, height=400)

pn.Column(dm, btn)
```

```python
random.seed(0)
baseline = []

n_mice = 4
base_mice = grouped['3months_19042021'].copy()

random.shuffle(base_mice)

images = []
for animal in base_mice[:n_mice]:
    young_med_frame = compute_frame_avg(animal, thresh=13, rears=True)
    img = hv.Image(young_med_frame, bounds=(0, 0, 80, 80)).opts(cmap='cubehelix', invert_yaxis=True)
    pp = pp_factory()
    dm = hv.DynamicMap(pp.plot_point, streams=[hv.streams.Tap(source=img)])
    images.append(img * dm.opts(framewise=True))
    baseline.append(pp)
reduce(add, images)
```

```python
def whitelist_subset(subsets, d):
    return keyfilter(lambda k: any(s in k for s in subsets), d)
```
