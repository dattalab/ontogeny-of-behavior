```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_11/ontogeny_males_mouse_area_df_v00.parquet')
```

```python
df['quant_0.5'].hist()
```

```python
df2 = df.reset_index()
tmp = df2.groupby('age').mean(numeric_only=True).reset_index()
```

```python
tmp['log_quant_0.5'] = np.log(tmp['quant_0.5'])
tmp['log_age'] = np.log(tmp['age'])
```

```python
ax = tmp.plot(x='log_age', y='quant_0.5')
ax.set(ylabel="Area (px^2)")
```

```python
ax = tmp.plot(x='age', y='log_quant_0.5')
ax.set(ylabel="Log Area (px^2)")
```
