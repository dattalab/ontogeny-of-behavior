```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import colorcet as cc
from statsmodels.gam.smooth_basis import BSplines
from aging.plotting import (
    figure,
    format_plots,
    add_identity,
    legend,
    PlotConfig,
    save_factory,
)
```

```python
format_plots()
c = PlotConfig()
saver = save_factory(c.save_path / "bio_age_model", tight_layout=False)
```

```python
splines = BSplines(np.linspace(0, 1, 500), 6, 3, include_intercept=True)
```

```python
fig, ax = plt.subplots(6, 1, figsize=(0.5, 1.3), sharex=True, sharey=True)
for a, spline, c in zip(ax.flat, splines.basis.T, cc.glasbey[1:]):
    a.plot(spline, c=c)
ax[2].set(ylabel="Spline weight")
ax[-1].set(xlabel="Age (weeks)", xticks=[], ylim=(0, 1), yticks=[])
sns.despine()
saver(fig, "spline-basis-schematic");
```
