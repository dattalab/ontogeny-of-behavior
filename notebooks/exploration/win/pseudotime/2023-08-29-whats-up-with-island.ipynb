```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
outliers = pd.read_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_05/male_long_outlier_sessions.parquet')

outlier_uuids = outliers.index.get_level_values('uuid')
```

```python
df = pd.read_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_05/longtogeny_males_syllable_df_v01.parquet')
```

```python
all_avgs = df.groupby('uuid', sort=False).mean(numeric_only=True)
```

```python
outlier_avgs = df[df['uuid'].isin(outlier_uuids)].groupby('uuid', sort=False).mean(numeric_only=True)
other_avgs = df[~df['uuid'].isin(outlier_uuids)].groupby('uuid', sort=False).mean(numeric_only=True)
```

```python
sns.histplot(outlier_avgs['true_depth'], stat='density')
sns.histplot(other_avgs['true_depth'], color='C1', stat='density')
```

```python
sns.histplot(outlier_avgs['velocity_2d_mm'], stat='density', label='Outlier')
sns.histplot(other_avgs['velocity_2d_mm'], color='C1', stat='density', label='Other')
plt.legend()
```

```python
sns.histplot(outlier_avgs['rescaled_width'], stat='density')
sns.histplot(other_avgs['rescaled_width'], color='C1', stat='density')
```

```python
sns.histplot(outlier_avgs['rescaled_area'], stat='density')
sns.histplot(other_avgs['rescaled_area'], color='C1', stat='density')
```

```python
sns.histplot(outlier_avgs['recon_width'], stat='density')
sns.histplot(other_avgs['recon_width'], color='C1', stat='density')
```

```python
sns.histplot(all_avgs['velocity_2d_mm'], stat='density')
```
