```python
import pandas as pd

from aging.behavior.pseudotime import compute_pseudotime, pseudotime_springplot
```

```python
df = pd.read_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_05/ontogeny_females_clean_v2.parquet')
```

```python
def remove_crl(df):
    return df[~df.index.get_level_values('session_name').str.contains("CRL")]
```

```python
pseudo_df = compute_pseudotime(df, filter_fun=remove_crl, k_neigh=6)
```

```python
graph, pos, colors = pseudotime_springplot(df, filter_fun=remove_crl, k_neigh=6, seed=1)
```

```python
df.join(pseudo_df).to_parquet('/n/groups/datta/win/longtogeny/data/ontogeny/version_05/ontogeny_females_clean_pseudotime_v02.parquet')
```
