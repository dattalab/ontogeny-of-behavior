```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import colorcet as cc
import numpy as np
from aging.plotting import figure, save_factory, PlotConfig, format_plots
```

```python
format_plots()
c = PlotConfig()
saver = save_factory(c.save_path / "lab-meeting")
```

```python
arr = np.arange(16 * 22).reshape(22, 16).T
```

```python
brightness = arr // 16
```

```python
cmap1 = cc.cm.glasbey(np.linspace(0, 1, 256))
cmap2 = cc.cm.glasbey_dark(np.linspace(0, 1, 256))
```

```python
all_colors = np.vstack((cmap1, cmap2))
```

```python
mymap = mcolors.LinearSegmentedColormap.from_list('mymap', all_colors, N=512)
```

```python
fig = figure(2, 1.5)
norm = mpl.colors.Normalize(vmin=4, vmax=90)
sns.heatmap(arr, cmap=mymap, cbar=False)
ax = fig.gca()
ax.set(xticks=[], yticks=[], xlabel="Age", ylabel="Mouse", title="Ontogeny")
fig.colorbar(mpl.cm.ScalarMappable(norm, cmap='binary'), ax=ax, label='Age (weeks)')
saver(fig, 'ontogeny-recording-schematic');
```

```python
long_arr = np.tile(np.arange(16), (22, 1)).T
```

```python
fig = figure(2, 1.5)
norm = mpl.colors.Normalize(vmin=4, vmax=90)
sns.heatmap(long_arr, cmap=mymap, vmax=512, cbar=False)
ax = fig.gca()
ax.set(xticks=[], yticks=[], xlabel="Age", ylabel="Mouse", title="Longtogeny")
fig.colorbar(mpl.cm.ScalarMappable(norm, cmap='binary'), ax=ax, label='Age (weeks)')
saver(fig, 'longtogeny-recording-schematic');
```
