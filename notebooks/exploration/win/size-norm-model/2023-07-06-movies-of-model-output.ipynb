```python
import h5py
import random
import numpy as np
from pathlib import Path
from tqdm.auto import tqdm
from ipywidgets import Video
from aging.video import write_movie
from aging.size_norm.data import clean
```

```python
files = sorted(Path('/n/groups/datta/win/longtogeny/data/ontogeny/version_03/all_data_pca/Ontogeny_males').glob('*.h5'))
```

```python
len(files)
```

```python
files[:5]
```

```python
def get_area(file):
    with h5py.File(file, 'r') as h5f:
        return np.nanmedian(h5f['scalars/area_mm'])
```

```python
# sort files by area
sorted_files = sorted(files, key=get_area)
```

```python
indices = list(range(len(files)))
```

```python
random.shuffle(indices)
```

```python
n_frames = 3000
for idx in tqdm(indices):
    with h5py.File(sorted_files[idx], 'r') as h5f:
        frames = np.array(list(map(clean, h5f['frames'][:n_frames])))
        recon_frames = h5f['win_size_norm_frames_v2'][:n_frames]
    _input = np.concatenate((frames, recon_frames), axis=2)
    write_movie(f'/n/groups/datta/win/longtogeny/data/video/male_ontogeny_example_file_{idx:03}.mp4', _input)
```
