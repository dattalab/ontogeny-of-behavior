```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exponential_scaling(start_value, max_value, time_to_max, start_time, current_time):
    # Calculate the scaling factor based on the time elapsed
    scaling_factor = np.log(max_value / start_value) / time_to_max
    
    # Calculate the scaled value at the current time
    scaled_value = start_value * np.exp(scaling_factor * (current_time - start_time))

    # Make sure the scaled value does not exceed the maximum value
    scaled_value = np.clip(scaled_value, 0, max_value)
    if isinstance(current_time, np.ndarray):
        scaled_value[current_time < start_time] = 0
    else:
        scaled_value *= (current_time >= start_time)

    return scaled_value
```

```python
vals = exponential_scaling(1e-6, 1, 99, 0, np.arange(100))
```

```python
plt.plot(vals)
```
